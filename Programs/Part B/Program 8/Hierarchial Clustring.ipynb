```python
import numpy as np 
import pandas as pd
import seaborn as sb
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('advertising.csv')
df.drop(['Unnamed: 0','TV','radio'],inplace=True,axis=1)
df.head()
```

```python
plt.title("Dendograms - AGNES")
dend = shc.dendrogram(shc.linkage(df, method='ward',metric="euclidean"))
```

```python
from sklearn.cluster import AgglomerativeClustering

clustering_model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
clustering_model.fit(df)
```

```python
data_labels = clustering_model.labels_
sb.scatterplot(x='newspaper', 
                y='sales', 
                data=df, 
                hue=data_labels)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('AGNES')
```

```python
df = pd.read_csv("advertising.csv")
df.drop(['Unnamed: 0','TV','newspaper'],inplace=True,axis=1)
df.head()
```

```python
plt.title("Dendograms - DIANA")
dend = shc.dendrogram(shc.linkage(df, method='ward',metric="euclidean"),orientation='bottom')
```

```python
clustering_model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
clustering_model.fit(df)
```

```python
data_labels = clustering_model.labels_
sb.scatterplot(x='radio', 
                y='sales', 
                data=df, 
                hue=data_labels)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('DIANA')
```
