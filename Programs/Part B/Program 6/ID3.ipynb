```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder 
from sklearn.metrics import *
import seaborn as sb
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
df = pd.read_csv('playtennis.csv')
df.head()
```

```python
encoder = LabelEncoder()
df = df.apply(encoder.fit_transform)
```

```python
df
```

```python
X = df.drop(["Play Tennis"], axis=1)
Y = df["Play Tennis"]
```

```python
model_id3 = DecisionTreeClassifier(criterion = 'entropy' , max_depth = 2)

model_id3.fit(X,Y)
pred = model_id3.predict(X)
tree.plot_tree(model_id3)
```

```python
print(classification_report(Y , pred))
```

```python
cm = confusion_matrix(Y , pred)
df = pd.DataFrame(cm , range(2) , range(2))
```

```python
sb.heatmap(df , annot = True , annot_kws = {'size':22})
```

```python
model_cart = DecisionTreeClassifier(criterion = 'gini' , max_depth = 2)

model_cart.fit(X,Y)
pred = model_cart.predict(X)
tree.plot_tree(model_cart)
```

```python
print(classification_report(Y , pred))
```

```python
cm = confusion_matrix(Y , pred)
df = pd.DataFrame(cm , range(2) , range(2))
```

```python
sb.heatmap(df , annot = True , annot_kws = {'size':22})
```
