```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error
import numpy as np
df=pd.read_csv('advertising.csv')
print(df.head())
```

```python
graphSheet = plt.figure(figsize=(20, 20))
graphSheet.add_subplot(3,3,1)
plt.scatter(df['TV'],df['sales'])
plt.xlabel("Money Spent on TV")
plt.ylabel("sales")
```

```python
graphSheet = plt.figure(figsize=(20, 20))
graphSheet.add_subplot(3,3,1)
plt.scatter(df['radio'],df['sales'],c='orange')
plt.xlabel("Money Spent on Radio")
plt.ylabel("sales")
```

```python
graphSheet = plt.figure(figsize=(20, 20))
graphSheet.add_subplot(3,3,1)
plt.scatter(df['newspaper'],df['sales'],c='teal')
plt.xlabel("Money Spent on Newspaper")
plt.ylabel("sales")
```

```python
def linReg(x,y):
    print(x+" vs " + y)
    print()
    # train_test_split with 70-30 ratio
    
    x_train,x_test,y_train,y_test=train_test_split(df[x],df[y],test_size=0.3)
    x_train = x_train.to_numpy().reshape(-1,1)
    x_test = x_test.to_numpy().reshape(-1,1)
    y_train = y_train.to_numpy().reshape(-1,1)
    y_test = y_test.to_numpy().reshape(-1,1)
    
    # 
    
    #Initializing and fitting LinearRegression model
    
    lr = LinearRegression()
    lr.fit(x_train,y_train)
    
    #Printing Coeff and Intercept values
    
    print("Coeff=",lr.coef_[0][0],"\nIntercept=",lr.intercept_[0])
    pred = lr.predict(x_test)
    
    #Prining line of regression
    
    print("The linear model of {} versus {} is: Y = {:.3} + {:.2}X".format(x,y,lr.intercept_[0], lr.coef_[0][0]))
    
    #Finding Root Mean Sqaured Error
    
    rmse=np.sqrt(mean_squared_error(y_test,pred))
    
    #Finiding Mean Absolute Error
    
    mae = mean_absolute_error(y_test,pred)
    
    #Finding Mean Squared Error
    
    mse = mean_squared_error(y_test,pred)
    
    #Printing all error matrics
    
    print("Root Mean Square Srror= {}\nMean Absolute Error = {}\nMean Squared Error = {}".format(rmse,mae,mse))
    
    #Plotting the final Graph
    
    plt.scatter(x_train,y_train)
    plt.scatter(x_test,y_test)
    plt.xlabel("Money Spent on "+x)
    plt.ylabel(y)
    plt.title(x+" vs "+y)
    plt.plot(x_test,pred,c='black')
```

```python
linReg('TV', 'sales')
```

```python
linReg('radio', 'sales')
```

```python
linReg('newspaper' , 'sales')
```

## Inference

Based on the output, we can see that there is a positive linear relationship between each advertising medium (TV, radio, newspaper) and sales. This is indicated by the positive coefficients of the linear regression models.

However, the strength of the relationship seems to vary between the mediums. The coefficient for TV is the smallest at 0.048, while the coefficient for radio is the largest at 0.193. This suggests that the money spent on radio advertising has a stronger impact on sales than the money spent on TV or newspaper advertising.

We can also see that the RMSE and MSE for newspaper advertising are higher than those for TV and radio advertising, indicating that the linear regression model for newspaper advertising may not be as accurate as those for TV and radio advertising. This may suggest that the relationship between newspaper advertising and sales is not as strong as the relationships between TV or radio advertising and sales.
